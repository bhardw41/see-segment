## Generational Fitness
This section will experiment with the toy data sets. 
Here will create plots that demonstrate how the best fitness score changes over with each
generation. We expect the fitness score to monotonically decrease.

```python
# Create data sets
%run dataset_script.py --filename-tail=concern_trial_3 --num-gen=30 --pop-size=30
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob
from functools import reduce

# We loop because we expect multpile data files to have been 
# generated (from running trials on the HPCC)
datasets_names = ['Moons Dataset', 'Circles Dataset', 'Linearly Separable Dataset']
num_datasets = len(datasets_names)
dataframes = []

for ds_name in datasets_names:
    data_files = glob.glob("{}*".format(ds_name))
    #print(data_files)
    # extract dataframes from all data files
    ds_dataframes = list(map(lambda file: pd.read_csv(file, header=None), data_files))
    # condense dataframes' fitness columns based on generation number
    combined_df = reduce(lambda df_1, df_2: pd.merge(left=df_1, right=df_2, how='left', on=0), ds_dataframes)
    # attach average fitness as a column
    combined_df['avg'] = combined_df.iloc[:, 1:].mean(axis=1)
    combined_df.rename(columns={0: 'generation'})
    dataframes.append(combined_df)

#fig, axes = plt.subplots(nrows=1, ncols=3)

# Make loop

# Init trials matrix

trial_count = len(glob.glob("{}_fitness_concern*".format(datasets_names[0])))

trial_matrix = np.empty((trial_count,num_datasets),dtype='object')

for i, ds in enumerate(datasets_names):
    files = glob.glob("{}_fitness_concern*".format(ds))
    trial_matrix[:,i] = files

fig, axes = plt.subplots(nrows=num_datasets,ncols=trial_count)

for i, row in enumerate(trial_matrix):
    df_1 = pd.read_csv(row[0], header=None)
    df_2 = pd.read_csv(row[1], header=None)
    df_3 = pd.read_csv(row[2], header=None)
    dfs = [df_1, df_2, df_3]

    combined_dfs = reduce(lambda df_1, df_2: pd.merge(left=df_1, right=df_2, how='left', on=0), dfs)

    # Format Columns
    combined_dfs = combined_dfs.rename(columns={0:'Generation', '1_x': 'Moons', '1_y': 'Circles', 1: 'Linearly Separable'})

    combined_dfs = combined_dfs.set_index('Generation')

    combined_dfs.plot.line(
        title="Trial {}".format(i),
        subplots=True,
        ax=axes[:,i]
    )
```
