## Plot Generation vs Fitness Score

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob
from functools import reduce

# We loop because we expect multpile data files to have been
# generated (from running trials on the HPCC)
datasets_names = ["Moons", "Circles", "Linearly Separable"]
num_datasets = len(datasets_names)

# Init trials matrix
num_trials = len(glob.glob("{}_fitness_trial_*.csv".format(datasets_names[0])))

trial_matrix = np.empty((num_trials, num_datasets), dtype="object")

for i, ds in enumerate(datasets_names):
    files = sorted(glob.glob("{}_fitness_trial_*.csv".format(ds)))
    trial_matrix[:, i] = files

fig, axes = plt.subplots(nrows=num_datasets, ncols=num_trials, figsize=(50, 20))

trial_dataframes = []

# Use min_max to format axes
min_max = np.zeros((num_datasets, 2))
min_max[:, 0] = np.ones((1, num_datasets))

for i, trial in enumerate(trial_matrix):
    # Extract trial's data frames
    df_1 = pd.read_csv(trial[0], header=None)
    df_2 = pd.read_csv(trial[1], header=None)
    df_3 = pd.read_csv(trial[2], header=None)
    dfs = [df_1, df_2, df_3]

    # Combine data frames
    combined_dfs = reduce(
        lambda df_1, df_2: pd.merge(left=df_1, right=df_2, how="left", on=0), dfs
    )

    # Format columns and store trial data frame
    combined_dfs = combined_dfs.rename(
        columns={
            0: "Generation",
            "1_x": "Moons",
            "1_y": "Circles",
            1: "Linearly Separable",
        }
    )
    combined_dfs = combined_dfs.set_index("Generation")

    # TODO Iterate num_datasets
    current_min = combined_dfs["Moons"].min()
    current_max = combined_dfs["Moons"].max()
    if current_min < min_max[0][0]:
        min_max[0][0] = current_min
    if current_max > min_max[0][1]:
        min_max[0][1] = current_max

    current_min = combined_dfs["Circles"].min()
    current_max = combined_dfs["Circles"].max()
    if current_min < min_max[1][0]:
        min_max[1][0] = current_min
    if current_max > min_max[1][1]:
        min_max[1][1] = current_max

    current_min = combined_dfs["Linearly Separable"].min()
    current_max = combined_dfs["Linearly Separable"].max()
    if current_min < min_max[2][0]:
        min_max[2][0] = current_min
    if current_max > min_max[2][1]:
        min_max[2][1] = current_max

    trial_dataframes.append(combined_dfs)

    # Add trial titles
    axes[0][i].set_title("Trial {}".format(i + 1), {"fontsize": 25})


for i, trial in enumerate(trial_dataframes):
    # Plot trial
    a = trial.plot.line(fontsize=20, subplots=True, ax=axes[:, i], lw=8)

    # Format axis labels and set axes limits
    for j in range(num_datasets):
        a[j].set_ylabel("Fitness Score", {"fontsize": 20})
        a[j].set_xlabel("Generation", {"fontsize": 20})
        #a[j].xaxis.label.set_visible(False)
        a[j].set_ylim(min_max[j] - [0.009, -0.02])
        a[j].legend(loc=2, prop={"size": 20})

fig.tight_layout()
````
