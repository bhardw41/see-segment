```python
import pandas as pd
```

```python
df = pd.read_csv('OnlineNewsPopularity.csv')
```

Extract predicitive columns

```python
df_predictive = df.iloc[:, 2:-1]
predictive_traits = df_predictive.to_numpy()
```

```python
df_labels = df.iloc[:, -1]
labels = df_labels.to_numpy()
# 1 means popular
# 0 means not popular
labels = list(map(lambda num_shares: 1 if num_shares >= 1400 else 0, labels))
labels
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import RandomForestClassifier

classifiers = [
    AdaBoostClassifier(),
    KNeighborsClassifier(),
    SVC(),
    RandomForestClassifier(),
]
classifiers_names = ["Ada Boost", "KNN", "SVC", "Random Forest"]
```

```python
%%time

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np

# Number of iterations used is 10?

standard_scaler = StandardScaler()
predictive_traits = standard_scaler.fit_transform(predictive_traits)

random_states = [0, 21, 42, 313, 5525, 8599, 6490, 304737, 7231604, 7231655]

evaluation_matrix = np.zeros((len(random_states), len(classifiers)))

for i, random_state in enumerate(random_states):
    training_X, validation_X, training_y, validation_y = train_test_split(
        predictive_traits, labels, random_state=random_state, test_size=0.3
    )

    scores = np.zeros(len(classifiers))
    for j, clf in enumerate(classifiers):
        clf.fit(training_X, training_y)
        scores[j] = clf.score(validation_X, validation_y)
        print(f"Done with {classifiers_names[j]}")

    evaluation_matrix[i, :] = scores
    print(f"Done with iteration {i}")
```

```python
import pandas as pd

df = pd.DataFrame(
    evaluation_matrix,
    columns=classifiers_names,
    index=[f"Split with random-state: {rs}" for rs in random_states],
)
df.style.set_caption("Scores using default hyperparameters")
```
